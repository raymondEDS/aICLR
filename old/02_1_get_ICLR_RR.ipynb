```python
import openreview
import pandas as pd
import datetime as date
import tqdm
```

```python
df_submissions = pd.read_csv('./data/submissions/df_submissions.csv')
client = openreview.Client(baseurl='https://api.openreview.net')
```

```python
#takes about 10 minutes
list_all_notes = [client.get_notes(forum =forum) for forum in tqdm.tqdm(df_submissions.forum.tolist())]
```

```python
list_all_notes = [item for sublist in list_all_notes for item in sublist]
list_all_notes = [note.__dict__ for note in list_all_notes]
```

```python
df_all_notes = pd.DataFrame(list_all_notes)
df_all_notes = pd.concat([df_all_notes, df_all_notes.content.apply(pd.Series)],axis = 1)
df_all_notes = df_all_notes.drop(columns=['content'])
```

```python
def find_rr(row):
    row = row['invitation']
    row = row.lower()
    if (row.find('official') != -1):
        return 1
    else:
        return 0
    
def find_type(row):
    row = row['invitation']
    out = (row.rsplit('/',2)[-2:][0] + "_"+row.rsplit('/',2)[-2:][1]).lower()
    return out
    
def find_meta_review(row):
    row = row['invitation']
    row = row.lower()
    if (row.find('meta') != -1):
        return 1
    else:
        return 0

    

def find_year(row):
    row = row['invitation']
    x = row.split('/')
    return x[1]
    
    
#df_all_notes.invitation.str.rsplit(pat='/', n = 2,expand=True)

tqdm.tqdm.pandas()
df_all_notes['rr_flag'] = df_all_notes.progress_apply(lambda x: find_rr(x),axis=1)
df_all_notes['meta_flag'] = df_all_notes.progress_apply(lambda x: find_meta_review(x),axis=1)
df_all_notes['year'] = df_all_notes.progress_apply(lambda x: find_year(x),axis=1)
```

```python
df_all_notes.to_csv('./data/submissions/df_all_notes.csv',index=False)
```
