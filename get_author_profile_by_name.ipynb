```python
#import the OPENREVIEW API

#get authors
#Publication, name, ID, email, gender, role

import openreview
import pandas as pd
import datetime as date
import tqdm

df_authors = pd.read_csv('data/authors.csv')
df_authors = df_authors.drop('race',axis=1)
client = openreview.Client(baseurl='https://api.openreview.net')
```

```python
df_authors

print(df_authors.groupby(['year'],dropna=False)['year'].count())
```

```python
out = []
for index, row in tqdm.tqdm(df_authors.iterrows()):
    for profile in client.search_profiles(first=row['First_Name'], last = row['Last_Name']):
        input_row = {}
        input_row['First_Name'] = row['First_Name']
        input_row['Last_Name'] = row['Last_Name']
        input_row['forum'] = row['forum']
        for key,content in profile.__dict__.items():
            if key == 'content':
                for key_content, content_content in content.items():
                    input_row[key_content] = content_content
            else:
                input_row[key] = content
        out.append(input_row)
        
    #if index >10:
    #    break
df_author_profile = pd.DataFrame(out)
```
