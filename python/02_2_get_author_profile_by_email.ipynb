```python
import openreview
import pandas as pd
import datetime as date
import tqdm
import json
import ast
import itertools
import matplotlib.pyplot as plt
import datetime  
df_ICLR_all_submission = pd.read_csv('../data/submissions/df_submissions.csv')
client = openreview.Client(baseurl='https://api.openreview.net')
```

```python
df_ICLR_all_submission['time_real'] = df_ICLR_all_submission.tcdate.apply(lambda x: datetime.datetime.fromtimestamp(x/ 1000).strftime('%Y'))
df_ICLR_2017_submission = df_ICLR_all_submission[df_ICLR_all_submission['time_real']=="2016"]
df_ICLR_2018_submission = df_ICLR_all_submission[df_ICLR_all_submission['time_real']=="2017"]
df_ICLR_2019_submission = df_ICLR_all_submission[df_ICLR_all_submission['time_real']=="2018"]
```

```python
list_2017_emails = df_ICLR_2017_submission.authorids.dropna().apply(lambda x: x.strip('][').split(', ')).tolist()
list_2017_emails = list(set(list(itertools.chain.from_iterable(list_2017_emails))))
set_2017_emails = set(list_2017_emails)
list_2017_emails= list(set_2017_emails)
list_2017_emails = [email.replace("'","") for email in list_2017_emails ]
len(list_2017_emails)
```

```python
list_2018_emails = df_ICLR_2018_submission.authorids.dropna().apply(lambda x: x.strip('][').split(', ')).tolist()
list_2018_emails = list(set(list(itertools.chain.from_iterable(list_2018_emails))))
set_2018_emails = set(list_2018_emails)
list_2018_emails= list(set_2018_emails)
list_2018_emails = [email.replace("'","") for email in list_2018_emails ]
len(list_2018_emails)
```

```python
list_2019_emails = df_ICLR_2019_submission.authorids.dropna().apply(lambda x: x.strip('][').split(', ')).tolist()
list_2019_emails = list(set(list(itertools.chain.from_iterable(list_2019_emails))))
set_2019_emails = set(list_2019_emails)
list_2019_emails= list(set_2019_emails)
list_2019_emails = [email.replace("'","") for email in list_2019_emails ]
len(list_2019_emails)
```

```python
out = []



for email, content in client.search_profiles(emails = list_2017_emails).items():
    input_row = {}
    input_row['year'] = 2017
    input_row['email'] = email
    for profile in content:
        #print(profile.__dict__.items())
        for key , profile_content in profile.__dict__.items():
            if key == 'content':
                #print(profile_content)
                for key_content, profile_content_content in profile_content.items():
                    input_row[key_content] = profile_content_content
            else:
                input_row[key] = profile_content
    out.append(input_row)


for email, content in client.search_profiles(emails = list_2018_emails).items():
    input_row = {}
    input_row['year'] = 2018
    input_row['email'] = email
    for profile in content:
        #print(profile.__dict__.items())
        for key , profile_content in profile.__dict__.items():
            if key == 'content':
                #print(profile_content)
                for key_content, profile_content_content in profile_content.items():
                    input_row[key_content] = profile_content_content
            else:
                input_row[key] = profile_content
    out.append(input_row)

    
for email, content in client.search_profiles(emails = list_2019_author_emails).items():
    input_row = {}
    input_row['year'] = 2019
    input_row['email'] = email
    for profile in content:
        #print(profile.__dict__.items())
        for key , profile_content in profile.__dict__.items():
            if key == 'content':
                #print(profile_content)
                for key_content, profile_content_content in profile_content.items():
                    input_row[key_content] = profile_content_content
            else:
                input_row[key] = profile_content
    out.append(input_row) 
out
```
