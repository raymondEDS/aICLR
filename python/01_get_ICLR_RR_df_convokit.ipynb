```python
import openreview
import pandas as pd
import numpy as np
import datetime as date
import os
import tqdm
import ast
import json
client = openreview.Client(baseurl='https://api.openreview.net')
```

```python
notes_all_2017 = client.get_all_notes(invitation = 'ICLR.cc/2017/conference/.*')
notes_all_2018 = client.get_all_notes(invitation = 'ICLR.cc/2018/Conference/.*')
notes_all_2019 = client.get_all_notes(invitation = 'ICLR.cc/2019/Conference/.*')
len(notes_all_2017)
len(notes_all_2018)
len(notes_all_2019)
```

```python
out = []
for note in tqdm.tqdm(notes_all_2017):
    if note.invitation.rsplit('/',2)[-2:] == ['official', 'review'] or note.invitation.rsplit('/',2)[-2:] == ['official', 'comment'] or note.invitation.rsplit('/',2)[-2:] == ['public', 'comment']:
        #print(note.invitation)
        input_row = {}
        input_row['year'] = 2017
        input_row['type'] = (note.invitation.rsplit('/',2)[-2:][0] + "_"+note.invitation.rsplit('/',2)[-2:][1]).lower()
        input_row['invitation'] = note.invitation
        input_row['readers'] = note.readers
        input_row['writers'] = note.writers
        input_row['signatures'] = note.signatures
        for key in note.content.keys():
            input_row[key] = note.content[key]
        input_row['id'] = note.id
        input_row['original'] = note.original
        input_row['number'] = note.number
        input_row['cdate'] = note.cdate
        input_row['tcdate'] = note.tcdate
        input_row['tmdate'] = note.tmdate
        input_row['ddate'] = note.ddate
        input_row['forum'] = note.forum
        input_row['referent'] = note.referent
        input_row['replyto'] = note.replyto
        input_row['nonreaders'] = note.nonreaders
        input_row['details'] = note.details
        #input_row['tauthor'] = note.tauthor
        out.append(input_row)

df_notes_2017 = pd.DataFrame(out)

out = []
for note in tqdm.tqdm(notes_all_2018):
    #print(note.invitation)
    if note.invitation.rsplit('/',1)[-1] == "Official_Comment" or note.invitation.rsplit('/',1)[-1] == "Official_Review":
         #print(note.invitation)
        #print(note.invitation)
        input_row = {}
        input_row['year'] =2018
        input_row['type'] = note.invitation.rsplit('/',1)[-1].lower()
        input_row['invitation'] = note.invitation
        input_row['readers'] = note.readers
        input_row['writers'] = note.writers
        input_row['signatures'] = note.signatures
        for key in note.content.keys():
            input_row[key] = note.content[key]
        input_row['id'] = note.id
        input_row['original'] = note.original
        input_row['number'] = note.number
        input_row['cdate'] = note.cdate
        input_row['tcdate'] = note.tcdate
        input_row['tmdate'] = note.tmdate
        input_row['ddate'] = note.ddate
        input_row['forum'] = note.forum
        input_row['referent'] = note.referent
        input_row['replyto'] = note.replyto
        input_row['nonreaders'] = note.nonreaders
        input_row['details'] = note.details
        out.append(input_row)
        
df_notes_2018 = pd.DataFrame(out)
out = []
for note in tqdm.tqdm(notes_all_2019):
    #print(note.invitation)
    if note.invitation.rsplit('/',1)[-1] == "Official_Comment" or note.invitation.rsplit('/',1)[-1] == "Official_Review":
         #print(note.invitation)
        #print(note.invitation)
        input_row = {}
        input_row['year'] =2019
        input_row['type'] = note.invitation.rsplit('/',1)[-1].lower()
        input_row['invitation'] = note.invitation
        input_row['readers'] = note.readers
        input_row['writers'] = note.writers
        input_row['signatures'] = note.signatures
        for key in note.content.keys():
            input_row[key] = note.content[key]
        input_row['id'] = note.id
        input_row['original'] = note.original
        input_row['number'] = note.number
        input_row['cdate'] = note.cdate
        input_row['tcdate'] = note.tcdate
        input_row['tmdate'] = note.tmdate
        input_row['ddate'] = note.ddate
        input_row['forum'] = note.forum
        input_row['referent'] = note.referent
        input_row['replyto'] = note.replyto
        input_row['nonreaders'] = note.nonreaders
        input_row['details'] = note.details
        #input_row['tauthor'] = note.tauthor
        out.append(input_row)

df_notes_2019 = pd.DataFrame(out)

df_all_notes = pd.concat([df_notes_2017, df_notes_2018,df_notes_2019])
df_all_notes
```

```python
df_review_notes = df_all_notes[df_all_notes["type"] == 'official_review']
df_review_comments = df_all_notes[df_all_notes["type"].isin(['official_comment','public_comment'])]
```

```python
df_review_rebuttal_pairs_all = pd.merge(df_review_notes, df_review_comments, how="left", left_on = 'id', right_on ='replyto')

df_review_rebuttal_pairs_all.to_csv('../data/model_inputs/df_sbert_rr_pairs_all.csv')
```

```python
#making convokit dataset
df_convokit = df_all_notes[df_all_notes["type"].isin(['official_comment','public_comment','official_review'])]
df_convokit = df_convokit[['id','tcdate','comment','review','signatures','replyto','forum','year']]
df_convokit['text'] = df_all_notes['comment'].fillna('')  + df_all_notes['review'].fillna('') 
df_convokit = df_convokit.drop(columns = ['comment','review'])
df_convokit = df_convokit[df_convokit['text']!='']
df_convokit.to_csv('../data/model_inputs/df_convokit.csv', index = False)
```
