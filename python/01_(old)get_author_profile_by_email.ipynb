```python
import openreview
import pandas as pd
import datetime as date
import tqdm
import json
import ast
import itertools
import matplotlib.pyplot as plt

df_ICLR_2019_blind_submissions = pd.read_csv('data/authors/ICLR_2019_blind_submissions.csv')
df_ICLR_2017_author = pd.read_csv('data/authors/ICLR_2017_author.csv')
df_ICLR_other_blind_submissions = pd.read_csv('data/authors/ICLR_other_blind_submissions.csv')
client = openreview.Client(baseurl='https://api.openreview.net')
```

```python
df_ICLR_2018_blind_submissions = df_ICLR_other_blind_submissions[df_ICLR_other_blind_submissions['year']==2018]


df_ICLR_2018_blind_submissions

```

```python
df_2019_content = pd.DataFrame(df_ICLR_2019_blind_submissions.content.apply(lambda x: ast.literal_eval(str(x))).tolist())

list_2019_author_emails=list(itertools.chain.from_iterable(df_2019_content.authorids.tolist()))

set_2019_author_emails = set(list_2019_author_emails)

list_2019_author_emails = list(set_2019_author_emails)

len(list_2019_author_emails)
```

```python
df_2019_content.shape
```

```python
df_2018_content = pd.DataFrame(df_ICLR_2018_blind_submissions.content.apply(lambda x: ast.literal_eval(str(x))).tolist())

df_2018_content.authorids.tolist()

list_2018_emails=list(itertools.chain.from_iterable(df_2018_content.authorids.tolist()))

set_2018_emails = set(list_2018_emails)

list_2018_emails= list(set_2018_emails)

len(list_2018_emails)
```

```python
series_2017_emails = df_ICLR_2017_author.authorids.apply(lambda x: '['+str(x).replace(' ',"") + ']')

series_2017_emails = series_2017_emails.apply(lambda x: str(x).strip('][').split(','))


list_2017_author_emails = list(set(list(itertools.chain.from_iterable(series_2017_emails.tolist()))))


list_2017_author_emails

len(list_2017_author_emails)
```

```python
#print(type(list_2017_author_emails))
#print(type(list_2019_author_emails))

email_list =  list_2017_author_emails + list_2018_emails + list_2019_author_emails
email_list
```

```python
out = []



for email, content in client.search_profiles(emails = list_2017_author_emails).items():
    input_row = {}
    input_row['year'] = 2017
    input_row['email'] = email
    for profile in content:
        #print(profile.__dict__.items())
        for key , profile_content in profile.__dict__.items():
            if key == 'content':
                #print(profile_content)
                for key_content, profile_content_content in profile_content.items():
                    input_row[key_content] = profile_content_content
            else:
                input_row[key] = profile_content
    out.append(input_row)


for email, content in client.search_profiles(emails = list_2018_emails).items():
    input_row = {}
    input_row['year'] = 2018
    input_row['email'] = email
    for profile in content:
        #print(profile.__dict__.items())
        for key , profile_content in profile.__dict__.items():
            if key == 'content':
                #print(profile_content)
                for key_content, profile_content_content in profile_content.items():
                    input_row[key_content] = profile_content_content
            else:
                input_row[key] = profile_content
    out.append(input_row)

    
for email, content in client.search_profiles(emails = list_2019_author_emails).items():
    input_row = {}
    input_row['year'] = 2019
    input_row['email'] = email
    for profile in content:
        #print(profile.__dict__.items())
        for key , profile_content in profile.__dict__.items():
            if key == 'content':
                #print(profile_content)
                for key_content, profile_content_content in profile_content.items():
                    input_row[key_content] = profile_content_content
            else:
                input_row[key] = profile_content
    out.append(input_row) 
out
```

```python
df_author_profiles = pd.DataFrame(out)
```

```python
df_author_profiles.replace({'m':'Male',
                                    'M':'Male',
                                    'male':'Male',
                                    None:'Unspecified',
                                    '':'Unspecified',
                                    'Not Specified':'Unspecified',
                                    'female':'Female'})

df_author_profiles.to_csv('./data/authors/df_author_profiles.csv')
```
