```python
import openreview
import pandas as pd
import datetime as date
import tqdm
import json
import ast
import itertools
import matplotlib.pyplot as plt
import requests, json
import multiprocessing
client = openreview.Client(baseurl='https://api.openreview.net')
```

```python
df_author_profiles = pd.read_csv('../data/authors/df_author_profiles.csv')
```

```python
def get_openAlex_json(str_first_name = 'raymond',str_middle_name = '',str_last_name = 'zhang'):
    search = "https://api.openalex.org/authors?filter=display_name.search:" + str_first_name + "%20" + str_last_name + "&page=1&per-page=200"
    response = requests.get(search)
    res = json.loads(response.text)
    return res

```

```python
lst_author_names_dict = df_author_profiles['names'].tolist()

lst_author_names_dict = [ast.literal_eval(l) for l in lst_author_names_dict]

lst_one_name_author = [l[0] for l in lst_author_names_dict if len(l) == 1 ]

lst_multi_name_author = [l for l in lst_author_names_dict if len(l) != 1 ]
    
lst_one_name_author

#type(lst_author_names_dict[0][0])
```

```python
out = []
for dict_author_name in tqdm.tqdm(lst_one_name_author):
    dict_input_row = {}
    dict_input_row['name'] = dict_author_name['first'] + ' ' + dict_author_name['last']
    dict_input_row['json'] = get_openAlex_json(dict_author_name['first'],dict_author_name['middle'],dict_author_name['last'])
    out.append(dict_input_row)
    break
    
df_out = pd.DataFrame(out)
```
